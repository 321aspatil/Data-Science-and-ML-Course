```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('banknotes.csv')
```

```python
df
```

```python
df.info()
```

### Separate input and output data

```python
#x=df.drop('Class',axis=1)
x=df[['Variance','Skewness','Curtosis','Entropy']]
y=df['Class']
```

### Explore the data

```python
sns.countplot(x=y);
```

```python
y.value_counts()# 0:genune notes 1:fake notes
```

##### Note :Decision tree doesnt require scaling of features

```python
sns.pairplot(df,hue='Class')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,stratify=y)#stratified sample gives representation from all the classses
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
y.value_counts()[0]/len(y)*100,y.value_counts()[1]/len(y)*100
```

```python
y_test.value_counts()[0]/len(y_test)*100,y_test.value_counts()[1]/len(y_test)*100
```

### Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(random_state=0)
```

```python
classifier.fit(x_train,y_train)
```

### Plot the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,16))
plot_tree(classifier,fontsize=7,filled=True,class_names=['O','F'],feature_names=x.columns,precision=2);
```

### Performance Evaluation

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred);
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
classifier.feature_importances_
```

```python
imp=classifier.feature_importances_
```

```python
plt.figure(figsize=(16,9))
plt.title('Feature Importance')
plt.grid()
plt.xlabel('Feature Name')
plt.ylabel('%Importance')
plt.bar(x.columns,imp*100,color='magenta')
```
