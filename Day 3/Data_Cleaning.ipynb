### Missing Values

```python
df.count()
```

```python
df.info()
```

```python
#columwise number of null values/missing values
df.isnull().sum()
```

```python
#Dropping null values rowsf
df.dropna()
```

```python
#filling missing values with a common value
df['marks'].fillna(60)
```

```python
#filling missing values of marks with average marks of all rows
df['marks'].fillna(df['marks'].mean())
```

```python
#filling missing values of marks with median marks of all rows
df['marks'].fillna(df['marks'].median())
```

```python
df['class']
```

```python
#filling the missing class values with mode value
df['class'].fillna(df['class'].mode()[0])
```

```python
df.fillna(method='pad')#the missing values are filled with the previous values of same column
```

```python
df.fillna(method='backfill',inplace=True)#the missing values ar filled with the next values of same column
```
