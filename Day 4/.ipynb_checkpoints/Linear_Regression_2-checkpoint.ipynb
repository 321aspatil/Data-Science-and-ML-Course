```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([[151],[174],[138],[186],[128],[136],[179],[163],[152],[131]])
y=np.array([63,81,56,91,47,57,76,72,62,48])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model=LinearRegression()
```

```python
lin_model.fit(x,y)
```

```python
plt.ylabel('Weight')
plt.xlabel('Height')
plt.scatter(x,y);
```

```python
lin_model.predict([[185]])
```

```python
lin_model.predict([[169]])
```

```python
lin_model.predict([[144]])
```

```python
y_pred=lin_model.predict(x)
```

```python
plt.title('Height vs. Weight')
plt.grid()
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x,y,color='r',label='Actual');
plt.plot(x,y_pred,color='g',label='Predicted');
plt.legend()
```

```python
#r2 value
lin_model.score(x,y)
```

```python
import numpy as np
#mae
np.mean(abs(y_pred-y))
```

```python
#mean absolute error in percentage
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_pred,y)
```
