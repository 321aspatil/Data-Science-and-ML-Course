```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
carsdf=pd.read_csv('mtcars.csv')
carsdf
```

```python
#separate into input and output
x=carsdf[['disp','hp','wt']]
y=carsdf['mpg']
```

```python
x.shape
```

```python
y.shape
```

```python
x.head()
```

```python
y.head()
```

```python
#Build the model
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
##prediction
'''disp=221 hp=102 wt=102'''
reg.predict([[221,102,2.91]])
```

```python
#accoracy r2 value
reg.score(x,y)
```
