```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.array([12,23,34,45,56])
y=np.array([15,21,37,42,59])
z=np.array([87,75,54,36,25])
```

```python
print(x)
print(y)
print(z)
```

### LinePlot

```python
plt.title('Line Plot x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,color='r',marker='o',mfc='black',ls='--',ms=10);
```

```python
plt.title('Line Plot x vs z')
plt.xlabel('x')
plt.ylabel('z')
plt.plot(x,z,color='g',marker='<',mfc='blue',lw=3)
```

```python
plt.plot(x,y,color='r',marker='o',mfc='black',ls='--',ms=10,label='2022');
plt.plot(x,z,color='g',marker='<',mfc='blue',lw=3,label='1922');
plt.legend()
plt.grid()
```

```python
import math
v=np.arange(0,4*math.pi,0.01)
w=np.sin(v)
u=np.cos(v)
plt.plot(v,w);
plt.plot(v,u);
```
