```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.array([12,23,34,45,56])
y=np.array([15,21,37,42,59])
z=np.array([87,75,54,36,25])
```

```python
import math
v=np.arange(0,4*math.pi,0.01)
w=np.sin(v)
u=np.cos(v)
plt.plot(v,w);
plt.plot(v,u);
```

### Subplot

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('Line Plot x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,color='r',marker='o',mfc='black',ls='--',ms=10);
plt.grid()
plt.subplot(2,2,2)
plt.title('Line Plot x vs z')
plt.xlabel('x')
plt.ylabel('z')
plt.plot(x,z,color='g',marker='<',mfc='blue',lw=3)
plt.grid()

plt.subplot(2,2,3)
v=np.arange(0,4*math.pi,0.01)
w=np.sin(v)
plt.plot(v,w);
plt.grid()
plt.subplot(2,2,4)
u=np.cos(v)
plt.plot(v,u);
plt.grid();
```

### Barplot

```python
courses=['Java','C','Python','R','Kotlin']
students=[67,70,74,45,50]
```

```python
plt.bar(courses,students,width=0.5,edgecolor='blue',ls='--',linewidth=2,alpha=0.5,color=['red','green','gold','silver','cyan']);
plt.xlabel('Course')
plt.ylabel('No. of Students')
plt.title('Student Enrollment')
#plt.grid()
```

### Boxplot

```python
help(plt.boxplot)
```

```python
marks=np.array([67,63,62,69,59,66,61,78,81])
```

```python
plt.boxplot(marks);
plt.grid()
```

```python
Q1=np.percentile(marks,25)
```

```python
Q3=np.percentile(marks,75)
```

```python
IQR=Q3-Q1
IQR
```

```python
upbound=Q3+1.5*IQR
lwbound=Q1-1.5*IQR
print('Upper bound=',upbound,'\nLower bound=',lwbound)
```
